```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.DataFrame()
df['input rate (Hz)'] = [10, 50, 100, 500, 1000, 10000, 50000]
df['count'] = [68, 171, 261, 1259, 2379, 23393, 118016]
df['time interval (s)'] = [10.21, 10.15, 10.17, 10.14, 10.03, 10.16, 10.25]
df['measured rate (Hz)'] = df['count']/df['time interval (s)']
df['measured to imput ratio'] = df['measured rate (Hz)']/df['input rate (Hz)']
```

```python
plt.figure(figsize=(10,6))
g = sns.lineplot(data=df, x='input rate (Hz)', y='measured to imput ratio', marker='o', dashes=False)
g.set(xscale='log')
g.grid('both')
```
